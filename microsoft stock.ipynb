## DATA ANALYSIS & VISUALIZATION ON MICROSOFT STOCK

### IMPORT MODULE

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### READ CSV FILE

```python
ms=pd.read_csv('D:/data visualization/Microsoft_Stock.csv')
```

### SHOWING DATASET

```python
pd.set_option('display.max_column' ,None)
pd.set_option('display.max_rows' ,None)
ms
```

### DATA CLEANING

```python
to_drop = ['Gain']

ms.drop(to_drop, inplace=True, axis=1)
```

### PRINT TOP 5 ROWS

```python
ms.head()
```

### PRINT BOTTOM 5 ROWS

```python
ms.tail()
```

### INFORMATION ABOUT DATASET

```python
ms.info()
```

```python
ms['Date']=pd.to_datetime(ms['Date'])
```

```python
ms['Date'][0]
```

### HIGH STOCK VALUE WITH DATE

```python
plt.figure(figsize=(20,12))
plt.plot(ms['Date'],ms['High'],color="r")
plt.title('Microsoft')
plt.xlabel('Date', fontsize=14)
plt.ylabel('High', fontsize=14)
plt.grid(True)
# (which="major", color='k', linestyle='-.', linewidth=0.5)
```

### SECURITY TRADED(VOLUME) ACCORDING TO DATE

```python
plt.figure(figsize=(20,12))
plt.plot(ms['Date'],ms['Volume'],color="mediumvioletred")
plt.title('Microsoft')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Volume', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
```

### INITIAL QUANTITY OF STOCK 

```python
plt.figure(figsize=(20,12))
plt.plot(ms['Date'],ms['Open'],color='c')
plt.title('Microsoft')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Open', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
```

### LOWEST PRICE OF STOCK ACCORDING TO DATE

```python
plt.figure(figsize=(20,12))
plt.plot(ms['Date'],ms['Low'],color='navy')
plt.title('Microsoft')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Low', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
```

### END OF TRADING SESSION ACCORDING TO DATE

```python
plt.figure(figsize=(20,12))
plt.plot(ms['Date'],ms['Close'],color='lime')
plt.title('Microsoft')
plt.xlabel('Year', fontsize=14)
plt.ylabel('Close', fontsize=14)
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
```
